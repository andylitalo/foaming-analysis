# Pressure and Temperature as Dry Ice Sublimates in VORANOL 360

Pressure and temperature plots along with some first-try analysis of solubility.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import medfilt

import solub
import plot


# user parameters
data_file = '../../EXPERIMENTS/Caltech/co2_p_in_parr/Data/20190408-0417_co2_30-8g_v360_25mL.csv'

# save parameters
save_folder = '../../EXPERIMENTS/Caltech/co2_p_in_parr/Figures/'

# plot parameters
ax_fs = 16
tk_fs = 12
t_fs = 20
lw = 2
ms = 4
save_prefix = '20190408_30-8g'
title = 'Dry Ice w V360 in Parr Reactor'
```

## Load Data

Load pressure, temperature, and time. Set negative values of pressure to zero. Median filter pressure to remove blips (result of power demand exceeding capacity due to LED readout).

```python
# load data
data = pd.read_csv(data_file, delimiter=',', header=0)
p = data['p (psi)']
T = data['T_thermo (C)']
t = data['t (s)']

# set negative pressures to zero
neg_p = p.values < 0
p.values[neg_p] = 0

# median filter
p = medfilt(p, kernel_size=11)
```

## Plot Data

We first plot early times.

```python
# early times
save_path = save_folder + save_prefix + '_p_T_early.pdf'
plot.plot_p_T_vs_t(p, T, t, save_plot=True, save_path=save_path, t_lim=[0,3000], title=title)
```

At first the temperature was not logged. Then we see a rapid drop in temperature from 20C to -35C in about 5 minutes. The temperature gradually warms up thereafter.

The pressure rises almost linearly, then slows aroudn 2000 s.

```python
# early times, semilog time
save_path = save_folder + save_prefix + '_p_T_semilogx.pdf'
plot.plot_p_T_vs_t(p, T, t, save_plot=True, save_path=save_path, log_t=True, t_lim=[20, 10000], title=title)
```

When time is plotted on a log scale, we see that the pressure grows linearly with log(t) between about 1000-4000 s. After, the pressure growth reaches a kink, as does temperature growth, possibly indicating complete sublimation of the dry ice.

```python
# full plot
save_path = save_folder + save_prefix + '_p_T_full.pdf'
plot.plot_p_T_vs_t(p, T, t, save_plot=True, save_path=save_path, title=title)
```

Full plot. We see that the temperature does not drift it warms up initially and the pressure slowly decays, appearing almost to plateau after 600,000 s. Note that 100,000 s ~ 1 day, and 600,000 ~ 1 week.

```python
# plot of pressure over entire time
save_path = save_folder + save_prefix + '_p_full.pdf'
plot.plot_p_vs_t(p, t, save_plot=True, save_path=save_path, p_lim=[620,700], title=title)
```

We now zoom in on the decay of the pressure and see that the decay had not completely stopped after 1 week. The diffusivity of polyol > 1E-10 m^2/s. Since the depth of the polyol is 1 cm, the time scale of diffusion is $$\tau_D > \frac{(0.01~m)^2}{(10^{-10}~m^2/s} = 10^6~s$$. Thus it is possible that diffusion has not finished completely, but likely that it is very close, so pressure loss near 800,000 s may be due to leaking.

```python
# plot of pressure over time on log-log scale
plot.plot_p_vs_t(p, t, p_lim=[620,700], log_p=True, log_t=True, t_lim=[20000,800000])
```

When we plot the pressure at late times (2-9 days) on a log-log plot, we see that the pressure decay is fairly linear, indicating a power-law decay. Near 800,000 s, the pressure decay levels off, possibly indicating a shift from pressure loss due to CO2 dissolving in the polyol to leaking.

```python
# plot of pressure over intermediate times
plot.plot_p_vs_t(p, t, p_lim=[630,670], log_p=True, t_lim=[100000,250000])
```

If we plot the log of pressure for intermediate times (1-3 days), the pressure decrease appears linear, indicating exponential decay.

## Fitting

```python
# select exponential decay regime [s]
exp_lo = 100000
exp_hi = 250000
exp_regime = (t > exp_lo) & (t < exp_hi)
t_exp = t[exp_regime]
p_exp = p[exp_regime]

# regression on log(p) vs. t for exponential fit
a_exp, b_exp = np.polyfit(t_exp, np.log(p_exp), 1)
t_exp_fit = np.linspace(exp_lo, exp_hi, 100)
p_exp_fit = np.exp(b_exp)*np.exp(a_exp*t_exp_fit)

ax = plot.plot_p_vs_t(p, t, t_lim=[exp_lo, exp_hi], log_p=True, p_lim=[640,660], title=title)
ax.semilogy(t_exp_fit, p_exp_fit, 'r--', label='p=%dexp(%.2gt)' % (np.exp(b_exp), a_exp))
plt.legend(loc='best')

plt.savefig(save_folder + save_prefix + '_p_exp_fit.pdf', bbox_inches="tight")
```

The pressure decay at intermediate times (1-3 days) appears to match exponential decay, although the uncertainty in the pressure is too great to be sure.

```python
# select power-law decay regime [s]
pwr_lo = 200000
pwr_hi = 700000
pwr_regime = (t > pwr_lo) & (t < pwr_hi)
t_pwr = t[pwr_regime]
p_pwr = p[pwr_regime]

# regression on log(p) vs. log(t) for power-law fit
a_pwr, b_pwr = np.polyfit(np.log(t_pwr), np.log(p_pwr), 1)
t_pwr_fit = np.linspace(pwr_lo, pwr_hi,100)
p_pwr_fit = np.exp(b_pwr)*t_pwr_fit**a_pwr

ax = plot.plot_p_vs_t(p, t, t_lim=[pwr_lo, pwr_hi], log_p=True, p_lim=[625,650], title=title)
ax.loglog(t_pwr_fit, p_pwr_fit, 'r--', label='p=%dt^(%.2g)' % (np.exp(b_pwr), a_pwr))
plt.legend(loc='best')

plt.savefig(save_folder + save_prefix + '_p_pwr_fit.pdf', bbox_inches="tight")
```

We see that the decay at late times (3-9 days) agrees with power-law decay, but the uncertainty is too great to be certain.

## Predict Solubility

```python
# density of VORANOL 360 acc. to SDS [g/mL]
rho_poly = 1.084 
# initial mass of dry ice [g] 
m_co2_0 = 30.82
# sublimation time [s]
t_subl = 60
# assume sublimation rate of 4 mg/s in open air [g/s]
rate_subl = 0.004
# volume of polyol [mL]
V_poly = 25
# volume of Parr reactor [mL]
V_parr = 240

# derived parameters
# mass of polyol
m_poly = V_poly * rho_poly
# mass of dry ice sealed in Parr reactor (subtract estimate of sublimation)
m_co2 = m_co2_0 - t_subl*rate_subl

p_gas = solub.m2p_v360(m_co2, m_poly, rho_poly=rho_poly, V=V_parr, p0=30E5)
print("Pressure in gas phase should be %.1f bar = %d psi." % (1E-5*p_gas, 14.5E-5*p_gas))
```

The pressure estimated based on the solubility of CO2 in VORANOL 360 measured by Dow is much higher than the measured pressure. This discrepancy is especially surprising because the above estimation assumes no swelling of the polyol, even though Di Caprio et al. (2016) indicates that the polyol will swell substantially as CO2 dissolves inside it. The estimated pressure in the gas phase is also fairly insensitive to the volume of polyol (changes by about 1 psi / mL of polyol change).

```python
# initial density of co2 in Parr reactor
rho_co2 = solub.interpolate_rho_co2(p=p_gas)
print(rho_co2)
```
