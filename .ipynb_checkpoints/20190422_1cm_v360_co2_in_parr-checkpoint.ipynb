# Pressure and Temperature as Dry Ice Sublimates in VORANOL 360

Pressure and temperature plots along with some first-try analysis of solubility.

```python
# import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.signal import medfilt

import solub
import plot


# user parameters
data_file = '../../EXPERIMENTS/co2_p_in_parr/Data/20190408-0422_co2_30-8g_v360_25mL.csv'

# save parameters
save_folder = '../../EXPERIMENTS/co2_p_in_parr/Figures/'

# plot parameters
ax_fs = 16
tk_fs = 12
t_fs = 20
lw = 2
ms = 4
save_prefix = '20190408_30-8g'
title = 'Dry Ice w V360 in Parr Reactor'
t_units = '(h)'
```

## Load Data

Load pressure, temperature, and time. Set negative values of pressure to zero. Median filter pressure to remove blips (result of power demand exceeding capacity due to LED readout).

```python
# load data
data = pd.read_csv(data_file, delimiter=',', header=0)
p = data['p (psi)']
T = data['T_thermo (C)']
t = data['t (s)'].values/3600 # s -> h

# set negative pressures to zero
neg_p = p.values < 0
p.values[neg_p] = 0

# median filter
p = medfilt(p, kernel_size=21)
```

## Plot Data

We first plot early times.

```python
# early times
save_path = save_folder + save_prefix + '_p_T_early.pdf'
plot.plot_p_T_vs_t(p, T, t, save_plot=True, save_path=save_path, t_lim=[0,1], t_units=t_units, 
                   title=title)
```

At first the temperature was not logged. Then we see a rapid drop in temperature from 20C to -35C in about 5 minutes. The temperature gradually warms up thereafter.

The pressure rises almost linearly, then slows aroudn 2000 s.

```python
# early times, semilog time
save_path = save_folder + save_prefix + '_p_T_semilogx.pdf'
plot.plot_p_T_vs_t(p, T, t, save_plot=True, save_path=save_path, log_t=True, t_lim=[0, 3],
                   t_units=t_units, title=title)
```

When time is plotted on a log scale, we see that the pressure grows linearly with log(t) between about 1000-4000 s. After, the pressure growth reaches a kink, as does temperature growth, possibly indicating complete sublimation of the dry ice.

```python
# full plot
save_path = save_folder + save_prefix + '_p_T_full.pdf'
plot.plot_p_T_vs_t(p, T, t, save_plot=True, save_path=save_path, t_units=t_units, title=title)
```

Full plot. We see that the temperature does not drift after it warms up initially. The pressure slowly decays, appearing almost to plateau after 200 hours. Note that 168 hours = 1 week.

```python
# plot of pressure over entire time
save_path = save_folder + save_prefix + '_p_full.pdf'
plot.plot_p_vs_t(p, t, save_plot=True, save_path=save_path, p_lim=[620,700], t_units=t_units,
                 title=title)
```

We now zoom in on the decay of the pressure and see that the decay had not completely stopped after 1 week. The diffusivity of polyol is $D_{v360} \in [10^{-10}~m^2/s, 10^{-7}~m^2/s]$ at 25 C (16 psi and 50 psi, respectively) (see "Diffusivity of CO2 in V360.xlsx" on Dow Sharepoint). Since the depth of the polyol is 1 cm, the time scale of diffusion is

\begin{align*}
\frac{(0.01~m)^2}{(10^{-7}~m^2/s)} &< \tau_D < \frac{(0.01~m)^2}{(10^{-10}~m^2/s)}\\
10^3~s < \tau_D < 10^6~s\\
3~h < \tau_D < 280~h\\
\end{align*}

Thus it is possible that diffusion has not finished completely, but likely that it is very close, so pressure loss after about 200 hours is predominantly due to leaking.

If we plot the log of pressure for early times (1-4 days), the pressure decrease does not appear linear. Instead we will try plotting on a log-log plot.

```python
# plot of pressure over intermediate times
plot.plot_p_vs_t(p, t, p_lim=[630,670], log_p=True, t_lim=[6,100], t_units=t_units)
```

When we plot the pressure at 1-4 days on a log-log plot, we see that the pressure decay is fairly linear, indicating a power-law decay. Near 9 days (200 hours), the pressure levels off, possibly indicating a shift from pressure loss due to CO2 dissolving in the polyol to leaking.

```python
# plot of pressure over time on log-log scale
plot.plot_p_vs_t(p, t, p_lim=[620,700], log_p=True, log_t=True, t_lim=[6,100], t_units=t_units)
```

```python
# plot of pressure over time on log-log scale
plot.plot_p_vs_t(p, t, p_lim=[620,640], log_p=True, t_lim=[100,350], t_units=t_units)
```

The decay is too slow after 100, 150 hours to determine the nature of the decay. We can effectively tree this range as a plateau with a decrease of 8 psi over the course of 240 hours (~10 days), amounting to **less than 1 psi lost per day.**

## Fitting

```python
# select exponential decay regime [h]
exp_lo = 30
exp_hi = 72
exp_regime = (t > exp_lo) & (t < exp_hi)
t_exp = t[exp_regime]
p_exp = p[exp_regime]

# regression on log(p) vs. t for exponential fit
a_exp, b_exp = np.polyfit(t_exp, np.log(p_exp), 1)
t_exp_fit = np.linspace(exp_lo, exp_hi, 100)
p_exp_fit = np.exp(b_exp)*np.exp(a_exp*t_exp_fit)

ax = plot.plot_p_vs_t(p, t, t_lim=[exp_lo, exp_hi], log_p=True, p_lim=[640,660],
                      t_units=t_units, title=title)
ax.semilogy(t_exp_fit, p_exp_fit, 'r--', label='p=%dexp(%.2gt)' % (np.exp(b_exp), a_exp))
plt.legend(loc='best')

plt.savefig(save_folder + save_prefix + '_p_exp_fit.pdf', bbox_inches="tight")
```

The pressure decay at intermediate times (1-3 days) appears to match exponential decay, although the uncertainty in the pressure is too great to be sure.

```python
# select power-law decay regime [h]
pwr_lo = 50
pwr_hi = 200
pwr_regime = (t > pwr_lo) & (t < pwr_hi)
t_pwr = t[pwr_regime]
p_pwr = p[pwr_regime]

# regression on log(p) vs. log(t) for power-law fit
a_pwr, b_pwr = np.polyfit(np.log(t_pwr), np.log(p_pwr), 1)
t_pwr_fit = np.linspace(pwr_lo, pwr_hi,100)
p_pwr_fit = np.exp(b_pwr)*t_pwr_fit**a_pwr

ax = plot.plot_p_vs_t(p, t, t_lim=[pwr_lo, pwr_hi], log_p=True, p_lim=[625,650],
                      t_units=t_units, title=title)
ax.loglog(t_pwr_fit, p_pwr_fit, 'r--', label='p=%dt^(%.2g)' % (np.exp(b_pwr), a_pwr))
plt.legend(loc='best')

plt.savefig(save_folder + save_prefix + '_p_pwr_fit.pdf', bbox_inches="tight")
```

We see that the decay at late times (3-9 days) agrees with power-law decay, but the uncertainty is too great to be certain.

We also want to try to fit the leaking regime (> 250 hours) to a leak profile. The leak profile assumes the gas is ideal ($PV = nRT$) and that the result from Bomelburg (1977) holds, *i.e.*, that $\dot{n} \propto p$, where $\dot{n}$ is the mass flow rate of gas through an orifice. Combining these, we get that

\begin{align*}
    \frac{\mathrm{d}p}{\mathrm{d}t} &= -kp\\
    \log(p/p_0) = -kt
    p &= p_0 e^{-kt}\\
\end{align*}

Thus, to fit to this behavior, we will perform a linear fit between $\log(p)$ and $t$.

```python
# select power-law decay regime [h]
leak_lo = 250
leak_hi = 350
leak_regime = (t > leak_lo) & (t < leak_hi)
t_leak = t[leak_regime]
p_leak = p[leak_regime]

# regression on log(p) vs. log(t) for power-law fit
a_leak, b_leak = np.polyfit(t_leak, np.log(p_leak), 1)
t_leak_fit = np.linspace(leak_lo, leak_hi,100)
p_0 = np.exp(b_leak)
p_leak_fit = p_0*np.exp(a_leak*t_leak_fit)

ax = plot.plot_p_vs_t(p, t, t_lim=[leak_lo, leak_hi], log_p=True, p_lim=[625,650],
                      t_units=t_units, title=title)
ax.loglog(t_leak_fit, p_leak_fit, 'r--', label='p={p_0:.1f}e^({a:.1f}t)'.format(p_0=p_0, a=a_leak))
plt.legend(loc='best')

plt.savefig(save_folder + save_prefix + '_p_leak_fit.pdf', bbox_inches="tight")
```

With this fit, we have estimated the leak rate throughout the experiment. We can therefore try to estimate when CO2 was no longer dissolved due to diffusion.

```python
# select data after pressure reached maximum (so all dry ice was roughly dissolved and leaking began to occur)
i_max = np.argmax(p)
p_decay = p[i_max:]
t_decay = t[i_max:]

# estimate loss of pressure due to leak
p_lost = p_0 - p_0*np.exp(a_leak*t_decay)

# pressure if there were no leak
p_no_leak = p_decay + p_lost

ax = plot.plot_p_vs_t(p_no_leak, t_decay, t_units=t_units, title=title)
```

But the corrected pressure does not plateau, rather it grows at some point, which is unphysical. Not sure what's going on.

## Predict Solubility

```python
# density of VORANOL 360 acc. to SDS [g/mL]
rho_poly = 1.084 
# initial mass of dry ice [g] 
m_co2_0 = 30.82
# sublimation time [s]
t_subl = 60
# assume sublimation rate of 4 mg/s in open air [g/s]
rate_subl = 0.004
# volume of polyol [mL]
V_poly = 25
# volume of Parr reactor [mL]
V_parr = 240

# derived parameters
# mass of polyol
m_poly = V_poly * rho_poly
# mass of dry ice sealed in Parr reactor (subtract estimate of sublimation)
m_co2 = m_co2_0 - t_subl*rate_subl

p_gas = solub.m2p_v360(m_co2, m_poly, rho_poly=rho_poly, V=V_parr, p0=30E5)
print("Pressure in gas phase should be %.1f bar = %d psi." % (1E-5*p_gas, 14.5E-5*p_gas))
```

The pressure estimated based on the solubility of CO2 in VORANOL 360 measured by Dow is much higher than the measured pressure. This discrepancy is especially surprising because the above estimation assumes no swelling of the polyol, even though Di Caprio et al. (2016) indicates that the polyol will swell substantially as CO2 dissolves inside it. The estimated pressure in the gas phase is also fairly insensitive to the volume of polyol (changes by about 1 psi / mL of polyol change).

```python
# initial density of co2 in Parr reactor
rho_co2 = solub.interpolate_rho_co2(p=p_gas)
print(rho_co2)
```
