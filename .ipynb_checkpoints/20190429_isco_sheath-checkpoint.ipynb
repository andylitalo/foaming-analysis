# Sheath Flow with ISCO Pumps

## Overview

ISCO 100 DM pump supplied blue-dyed glycerol through the inner stream and ISCO 260 D pump supplied glycerol through the outer stream to produce sheath flow in the microfluidic tee-junction device. This device was modified from previous versions after the glass inner capillary broken, having had the inner capillary replaced with a stainless steel inner capillary 2.4 cm long and with 400 um ID.

The following code analyzes images of these flows to compute the stream width and compare it to the stream width predicted by the Navier Stokes equations for 1D sheath flow in a cylinder.

```python
import numpy as np
import matplotlib.pyplot as plt

import improc
import flow
import data
import userinput as ui

### USER PARAMETERS ###
# desired radii of the inner stream [um]
r_i_arr = np.array([2.5, 5, 10, 15, 20, 25, 50, 75, 100, 125, 150, 200])
# pressure drop along observation capillary [bar]
dp_arr = np.arange(40, 105, 5)
# viscosity of fluid [Pa.s]
eta = 1.412
```

## Determine Flow Rates

We calculate flow rates to vary the radius of the inner stream at a fixed pressure drop fo 30 bar down the microfluidic channel.

```python
# pressure drop along observation capillary [bar]
dp = 30

# initialize arrays to store inner and outer stream flow rates [mL/min]
Q_i_arr = np.zeros_like(r_i_arr)
Q_o_arr = np.zeros_like(r_i_arr)

# calculate flow rates for each desired inner stream radius, given pressures at inlets (so tubing length set to 0)
for i in range(len(r_i_arr)):
    r_i = r_i_arr[i]
    Q_i_arr[i], Q_o_arr[i] = flow.get_flow_rates_ri_dp(eta, r_i, dp)

# print flow rates
ret = [print("r_i = %d um: Q_i = %.1f uL/min, Q_o = %.1f uL/min" \
             % (r_i_arr[i], Q_i_arr[i], Q_o_arr[i])) for i in range(len(r_i_arr))]
```

Assuming approximately 1 minute per experiment, the required supply of liquid is as follows.

```python
print("ISCO 100 DM requires %.1f mL of blue-dyed glycerol and ISCO 260 D requires %d mL of glycerol." \
     % (np.sum(Q_i_arr)/1000, np.sum(Q_o_arr)/1000))
```

Now vary the pressure drop with fixed radius of 30 um.

```python
# initialize arrays to store inner and outer stream flow rates [mL/min]
Q_i_arr = np.zeros_like(dp_arr)
Q_o_arr = np.zeros_like(dp_arr)

# calculate flow rates for each desired inner stream radius, given pressures at inlets (so tubing length set to 0)
for i in range(len(dp_arr)):
    dp = dp_arr[i]
    Q_i_arr[i], Q_o_arr[i] = flow.get_flow_rates_ri_dp(eta, r_i, dp)

# print flow rates
ret = [print("dp= %d bar: Q_i = %.1f uL/min, Q_o = %.1f uL/min" \
             % (dp_arr[i], Q_i_arr[i], Q_o_arr[i])) for i in range(len(dp_arr))]
```

Assuming approximately 1 minute per experiment, the required supply of liquid is as follows.

```python
print("ISCO 100 DM requires %.1f mL of blue-dyed glycerol and ISCO 260 D requires %d mL of glycerol." \
     % (np.sum(Q_i_arr)/1000, np.sum(Q_o_arr)/1000))
```

## Image Processing

We begin by loading the paths of all images we wish to analyze. Next we compute the stream width for each image. Finally we save the data.

If the data have already been processed, skip to the **"Analysis"** section.

```python
# Image processing parameters
data_folder = '..\\..\\EXPERIMENTS\\sheath_flow\\20190501_isco_sheath\\'
bf_file_name = 'bf_glyc_2.jpg'
im_hdr = 'blue_glyc_isco_*_*_*.jpg'
data_save_name = '20190501_stream_widths.csv'
columns = ['stream width [um]', 'standard deviation [um]']
um_per_pix = 0.63 # from looking at a sample image and counting pixels across ID of capillary

def parse_file_name(file_path, hdr='blue_glyc_isco_'):
    """Parse file name for these experiments."""
    # extract just the name of the file
    file_name = file_path[file_path.find(hdr):]
    # outer flow rate [uL/min]
    Q_o = int(file_name[len(hdr):len(hdr)+4])
    # inner flow rate [uL/min]
    Q_i = int(file_name[len(hdr)+5:len(hdr)+9])
    Q_i += float(file_name[len(hdr)+10])/10
    # trial number
    trial = int(file_name[-5])
    
    return Q_i, Q_o, trial    
    
# Check out the first image
im_path_list = data.get_filepaths(data_folder, im_hdr)
print(im_path_list[0])
im_sample = plt.imread(im_path_list[0])
plt.imshow(im_sample)
```

Now I will use k-means clustering to identify the inner stream and measure its width in microns (um).

```python
# get brightfield image
bf = plt.imread(data_folder + bf_file_name)
# Save stream widths computed using image processing
params = [um_per_pix, bf]
df = improc.proc_im_seq(im_path_list, improc.measure_stream_width, params, columns=columns)
df.head()
```

Having loaded the stream width results, I will compare them with the expected stream width.

```python
r_expected_list = []
dp_expected_list = []
p_i_expected_list = []
p_o_expected_list = []
Q_i_list = []
Q_o_list = []
for im_path in im_path_list:
    # Parse the file name to get flow parameters
    Q_i, Q_o, trial = parse_file_name(im_path)
    Q_i_list += [Q_i]
    Q_o_list += [Q_o]
    r_expected_list += [flow.get_inner_stream_radius(Q_i, Q_o)]
    p_i, p_o, p_inner_cap, p_obs_cap = flow.get_pressures(eta, Q_i, Q_o, l_inner_cap=2.4, r_inner_cap=200, 
                                                          l_tube_i=45, l_tube_o=50)
    dp_expected_list += [p_obs_cap]
    p_i_expected_list += [p_i]
    p_o_expeted_list += [p_o]
df['inner flow rate [uL/min]'] = np.array(Q_i_list)
dr['outer flow rate [uL/min]'] = np.array(Q_o_list)
df['stream width (expected) [um]'] = 2*np.array(r_expected_list)
df['Pressure drop (expected) [bar]'] = np.array(dp_expected_list)
df['ISCO 100 DM pressure (expected) [bar]'] = np.array(p_i_expected_list)
df['ISCO 260 D pressure (expected) [bar]'] = np.array(p_o_expected_list)
```
